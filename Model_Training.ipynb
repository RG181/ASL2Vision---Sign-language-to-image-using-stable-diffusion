```python
import numpy as np, os, random, json, shutil, requests
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import classification_report, confusion_matrix
import tensorflow as tf
from tensorflow.keras import layers, regularizers
from tensorflow.keras.applications import EfficientNetB3
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint
from tensorflow.keras.preprocessing import image as keras_image
from IPython.display import FileLink, Image as IPImage, display

# Reproducibility
SEED = 42
os.environ['PYTHONHASHSEED'] = str(SEED)
random.seed(SEED); np.random.seed(SEED); tf.random.set_seed(SEED)
print("GPU:", tf.config.list_physical_devices('GPU'))

# Auto-detect DATA_DIR
def find_data_dir():
    for root, dirs, files in os.walk('/kaggle/input'):
        letters = [d for d in dirs if d.upper() in 'ABCDEFGHIJKLMNOPQRSTUVWXYZ' and len(d)==1]
        if len(letters) >= 20:
            print(f"Data found: {root}  ({len(dirs)} classes)")
            return root
    raise FileNotFoundError("Add ASL Alphabet dataset via '+ Add Input'")

DATA_DIR = find_data_dir()

# ── Config (optimised for speed + accuracy) ──────────────────
IMG_SIZE        = 224    # Faster than 300, still excellent accuracy
BATCH_SIZE      = 64     # Larger batch = better GPU utilisation
VAL_SPLIT       = 0.15
PHASE1_EPOCHS   = 5
PHASE2_EPOCHS   = 3
PHASE1_LR       = 1e-3
PHASE2_LR       = 5e-6
UNFREEZE_LAYERS = 40
BEST_PATH       = "/kaggle/working/best_model.keras"
FINAL_PATH      = "/kaggle/working/asl_final.keras"

# ── Data generators ──────────────────────────────────────────
train_gen = ImageDataGenerator(
    validation_split=VAL_SPLIT,
    rotation_range=20, zoom_range=0.2,
    width_shift_range=0.15, height_shift_range=0.15,
    shear_range=0.1, brightness_range=[0.7, 1.3],
    fill_mode='nearest', horizontal_flip=False
).flow_from_directory(
    DATA_DIR, target_size=(IMG_SIZE, IMG_SIZE),
    batch_size=BATCH_SIZE, class_mode='categorical',
    subset='training', seed=SEED, shuffle=True
)

val_gen = ImageDataGenerator(
    validation_split=VAL_SPLIT
).flow_from_directory(
    DATA_DIR, target_size=(IMG_SIZE, IMG_SIZE),
    batch_size=BATCH_SIZE, class_mode='categorical',
    subset='validation', seed=SEED, shuffle=False
)

num_classes  = len(train_gen.class_indices)
class_labels = list(train_gen.class_indices.keys())
print(f"Train: {train_gen.samples} | Val: {val_gen.samples} | Classes: {num_classes}")

# ── Build model ──────────────────────────────────────────────
base = EfficientNetB3(
    input_shape=(IMG_SIZE, IMG_SIZE, 3),
    include_top=False, weights='imagenet'
)
base.trainable = False

inp = tf.keras.Input(shape=(IMG_SIZE, IMG_SIZE, 3))
x   = tf.keras.applications.efficientnet.preprocess_input(inp)
x   = base(x, training=False)
x   = layers.GlobalAveragePooling2D()(x)
x   = layers.BatchNormalization()(x)
x   = layers.Dense(512, activation='relu',
                   kernel_regularizer=regularizers.l2(1e-4))(x)
x   = layers.BatchNormalization()(x)
x   = layers.Dropout(0.5)(x)
x   = layers.Dense(256, activation='relu',
                   kernel_regularizer=regularizers.l2(1e-4))(x)
x   = layers.BatchNormalization()(x)
x   = layers.Dropout(0.3)(x)
out = layers.Dense(num_classes, activation='softmax')(x)
model = tf.keras.Model(inp, out)

model.compile(
    optimizer=tf.keras.optimizers.Adam(PHASE1_LR),
    loss=tf.keras.losses.CategoricalCrossentropy(label_smoothing=0.1),
    metrics=['accuracy',
             tf.keras.metrics.TopKCategoricalAccuracy(k=3, name='top3')]
)
print(f"Parameters: {model.count_params():,}")

# ── Callbacks ────────────────────────────────────────────────
def get_callbacks(patience_es=5, patience_lr=3):
    return [
        ModelCheckpoint(BEST_PATH, monitor='val_accuracy',
                        save_best_only=True, verbose=1),
        EarlyStopping(monitor='val_accuracy', patience=patience_es,
                      restore_best_weights=True, min_delta=0.001, verbose=1),
        ReduceLROnPlateau(monitor='val_loss', factor=0.3,
                          patience=patience_lr, min_lr=1e-8, verbose=1),
    ]

# ── Phase 1 ──────────────────────────────────────────────────
print("\n" + "="*50)
print("PHASE 1 — Training head (base frozen)")
print("="*50)
h1 = model.fit(
    train_gen, epochs=PHASE1_EPOCHS,
    validation_data=val_gen,
    callbacks=get_callbacks(5, 3), verbose=1
)
print(f"Phase 1 best val acc: {max(h1.history['val_accuracy'])*100:.2f}%")

# ── Phase 2 ──────────────────────────────────────────────────
print("\n" + "="*50)
print("PHASE 2 — Fine-tuning top layers")
print("="*50)
base.trainable = True
for layer in base.layers[:len(base.layers) - UNFREEZE_LAYERS]:
    layer.trainable = False

model.compile(
    optimizer=tf.keras.optimizers.Adam(PHASE2_LR),
    loss=tf.keras.losses.CategoricalCrossentropy(label_smoothing=0.1),
    metrics=['accuracy',
             tf.keras.metrics.TopKCategoricalAccuracy(k=3, name='top3')]
)
h2 = model.fit(
    train_gen, epochs=PHASE2_EPOCHS,
    validation_data=val_gen,
    callbacks=get_callbacks(4, 2), verbose=1
)
print(f"Phase 2 best val acc: {max(h2.history['val_accuracy'])*100:.2f}%")

# ── Plot training curves ─────────────────────────────────────
p1      = len(h1.history['accuracy'])
acc     = h1.history['accuracy']     + h2.history['accuracy']
val_acc = h1.history['val_accuracy'] + h2.history['val_accuracy']
loss    = h1.history['loss']         + h2.history['loss']
val_loss= h1.history['val_loss']     + h2.history['val_loss']
ep      = range(1, len(acc)+1)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))
ax1.plot(ep, acc,     'b-o', ms=3, label='Train')
ax1.plot(ep, val_acc, 'r-o', ms=3, label='Val')
ax1.axvline(x=p1, color='gray', linestyle='--', label='Fine-tune start')
ax1.set_title('Accuracy'); ax1.legend(); ax1.grid(alpha=0.3); ax1.set_ylim([0.4, 1.01])
ax2.plot(ep, loss,     'b-o', ms=3, label='Train')
ax2.plot(ep, val_loss, 'r-o', ms=3, label='Val')
ax2.axvline(x=p1, color='gray', linestyle='--')
ax2.set_title('Loss'); ax2.legend(); ax2.grid(alpha=0.3)
plt.tight_layout()
plt.savefig('/kaggle/working/training_curves.png', dpi=150)
plt.show()

gap = acc[-1] - val_acc[-1]
print(f"\nTrain: {acc[-1]*100:.2f}% | Val: {val_acc[-1]*100:.2f}% | Gap: {gap*100:.2f}%")
print("Status:", "GOOD" if gap < 0.03 else "MILD" if gap < 0.07 else "OVERFIT")

# ── Evaluate ─────────────────────────────────────────────────
best_model = tf.keras.models.load_model(BEST_PATH)
val_gen.reset()
loss_v, acc_v, top3_v = best_model.evaluate(val_gen, verbose=1)
print(f"\nFinal Val Accuracy : {acc_v*100:.2f}%")
print(f"Final Top-3 Accuracy: {top3_v*100:.2f}%")

val_gen.reset()
y_pred = np.argmax(best_model.predict(val_gen, verbose=1), axis=1)
y_true = val_gen.classes
print("\n", classification_report(y_true, y_pred,
                                   target_names=class_labels, digits=3))

cm = confusion_matrix(y_true, y_pred)
plt.figure(figsize=(22, 20))
sns.heatmap(cm, annot=True, fmt='d',
            xticklabels=class_labels, yticklabels=class_labels,
            cmap='Blues', linewidths=0.3)
plt.title('Confusion Matrix')
plt.tight_layout()
plt.savefig('/kaggle/working/confusion_matrix.png', dpi=150)
plt.show()

# ── Save ─────────────────────────────────────────────────────
best_model.save(FINAL_PATH)
label_map = {str(v): k for k, v in train_gen.class_indices.items()}
with open('/kaggle/working/class_labels.json', 'w') as f:
    json.dump(label_map, f, indent=2)
print(f"Model saved : {FINAL_PATH}")
print(f"Labels saved: /kaggle/working/class_labels.json")

# ── ZIP + download ────────────────────────────────────────────
BUNDLE = "/kaggle/working/bundle"
os.makedirs(BUNDLE, exist_ok=True)
for f in [FINAL_PATH,
          '/kaggle/working/class_labels.json',
          '/kaggle/working/training_curves.png',
          '/kaggle/working/confusion_matrix.png']:
    if os.path.exists(f):
        shutil.copy(f, BUNDLE)

shutil.make_archive('/kaggle/working/asl_model_complete', 'zip', BUNDLE)
mb = os.path.getsize('/kaggle/working/asl_model_complete.zip') / (1024*1024)
print(f"\nZIP ready: {mb:.1f} MB")
display(FileLink('asl_model_complete.zip',
                 result_html_prefix="<b>Click to download: </b>"))
print("OR: Right panel → Output → asl_model_complete.zip → Download")

# ── Predict + DALL-E ─────────────────────────────────────────
with open('/kaggle/working/class_labels.json') as f:
    lmap = {int(k): v for k, v in json.load(f).items()}

def predict_sign(img_path):
    img = keras_image.load_img(img_path, target_size=(IMG_SIZE, IMG_SIZE))
    arr = np.expand_dims(keras_image.img_to_array(img), axis=0)
    p   = best_model.predict(arr, verbose=0)[0]
    top = np.argsort(p)[::-1][:3]
    print("\nTop 3 predictions:")
    for i, idx in enumerate(top, 1):
        bar = '█' * int(p[idx] * 20)
        print(f"  {i}. {lmap[idx]:12s}  {p[idx]*100:.1f}%  {bar}")
    return lmap[top[0]]

def generate_image(word, key):
    r = requests.post(
        "https://api.openai.com/v1/images/generations",
        headers={"Authorization": f"Bearer {key}",
                 "Content-Type": "application/json"},
        json={"model": "dall-e-3",
              "prompt": f"Simple colorful illustration of '{word}', white background",
              "n": 1, "size": "1024x1024"}
    )
    res = r.json()
    if "error" in res:
        print("DALL-E error:", res["error"]["message"]); return
    import urllib.request
    out = f"/kaggle/working/dalle_{word}.png"
    urllib.request.urlretrieve(res["data"][0]["url"], out)
    display(IPImage(out, width=400))
    print(f"Generated image for: {word}")

# Auto-find a test image
test_img = None
for root, dirs, files in os.walk(DATA_DIR):
    for f in files:
        if f.lower().endswith(('.jpg', '.jpeg', '.png')):
            test_img = os.path.join(root, f); break
    if test_img: break

print(f"\nTesting with: {test_img}")
predicted = predict_sign(test_img)

from kaggle_secrets import UserSecretsClient
OPENAI_KEY = UserSecretsClient().get_secret("OPENAI_API_KEY")
generate_image(predicted, OPENAI_KEY)

print("\nALL DONE!")
```
